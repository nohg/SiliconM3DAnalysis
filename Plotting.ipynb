```python
NLTE3Dcorr_no_iosmet = np.load('../mdata/3DNLTE_abund_corr_no_iosmet_5wls.npy')
NLTE1Dcorr_no_iosmet = np.load('../mdata/1DNLTE_abund_corr_no_iosmet_5wls.npy')
LTE3Dcorr = np.load('../mdata/3DLTE_abund_corr_2404.npy')
```
